```python
#W6A1
A=list(map(int,input().split()))
def W6A1(A):
     B=[]
     for i in A:
          if i not in B:
               B.append(i)
     return B

print(W6A1(A))         
```

```python
#W6A2
A=list(map(int,input().split()))
def W6A2(A):
     B=[]
     bp=0

     for i in range(len(A)):
          for a in range(i+1):
               bp+=A[a]
          B.append(bp)
          bp=0
     return B 

print(W6A2(A))         
```

```python
#W6A9
A=input()
B=input()

def chuyen_dict(A):
     a=[]
     b={}
     for i in A:
          if i!="{" and i!="}" and i!="," and i!=":" and i!="'"and i!=" ":
               a.append(i)
     for i in range(0,len(a),2):
          b[a[i]]=int(a[i+1])
     return b     
     
def W6A9(A,B):
     A=chuyen_dict(A)
     B=chuyen_dict(B)
     for i in A:
          if i in B:B[i]=(B[i]+A[i])
          else:B[i]=A[i]
     return B

print(W6A9(A,B))   
```

```python
#W6A10
A=list(map(int,input().split()))
k=int(input())
def W6A10(A,k):
     Tuple=[]
     for i in range(len(A)):
          for a in range(i+1,len(A)):
               if A[i]+A[a]==k and (A[i],A[a])not in Tuple:Tuple.append((A[i],A[a]))
     Tuple.sort(key=lambda x:x[0]) 
     return Tuple
 
print(W6A10(A,k))      
```
