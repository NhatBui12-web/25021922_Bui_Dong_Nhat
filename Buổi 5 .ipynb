```python
#W5A3
def so_nguyen_to(n):
     tong=0
     if n<=1:return False
     else:
          for i in range(2,int(n**0.5)+1):
               if n%i==0:tong+=1
          if tong==0: return True
          else: return False 
```

```python
#W5A4
def so_hoan_hao(n):
     tong_uoc=0
     for i in range(1,round(n**0.5)+1):
          if n%i==0:
               if i!=(n/i):tong_uoc+=(i+(n/i))
               else:tong+=i
     if tong_uoc==2*n:return True
     else:return False
```

```python
#W5A5
def vi_tri(List:list,k:int):
     Vi_tri=-1
     for i in range(len(List)):
          if k==List[i]:Vi_tri=i
     return Vi_tri     
```

```python
#W5A6
def Giai_thua(n):
     giai_thua=1
     for i in range(1,n+1):giai_thua*=i
     return giai_thua
```

```python
#W5A7
def may_tinh(num1,operat,num2):
     if operat=="+":return num1+num2
     elif operat=="-":return num1-num2
     elif operat=="*":return num1*num2
     elif operat=="/":return round(num1/num2,2)
```

```python
#W5A8
def he_nhi_phan(n,N=[]):
     while n!=0:
          N.append(n%2)
          n//=2
     N.reverse() 
     return N
def Kc_Hamming(a,b):
     kc=0
     A=he_nhi_phan(a,N=[]) 
     B=he_nhi_phan(b,N=[])  
     if len(A)>=len(B):
          for i in range(len(A)-len(B)):
               B.insert(0,0)
          for i in range(len(B)) :
               if B[i]!=A[i]:kc+=1
     else:
          for i in range(len(B)-len(A)):
               A.insert(0,0)
          for i in range(len(B)):
               if B[i]!=A[i]:kc+=1
     return kc
```

```python
#W5A9
def tong_cac_chu_so(n):
     tong=0
     while n!=0:
          tong+=(n%10)
          n//=10
     return tong     
```
