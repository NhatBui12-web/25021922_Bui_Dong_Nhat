```python
#w7a1
def binary_search(arr,target):
     l,r=0,len(arr)-1
     while l<=r:
          gttg=(l+r)//2
          if arr[gttg]<target:l=gttg+1
          elif arr[gttg]>target:r=gttg-1
          else:return gttg
     return -1
```

```python
#w7a3
def bubble_sort(arr):
     for i in range(len(arr)):
          dss=True
          for y in range(0,len(arr)-i-1):
              if arr[y]>arr[y+1]:
                   arr[y],arr[y+1]=arr[y+1],arr[y]
                   dss=False
          if dss:break
     return arr

def selection_sort(arr):
     for i in range(len(arr)):
          min=i
          for y in range(i+1,len(arr)):
               if arr[min]>arr[y]:min=y
          arr[min],arr[i]=arr[i],arr[min]
     return arr
```

```python
#w7a5
def w7a5(arr,x):
     Dict={}
     cap=0

     for i in arr:
          if i in Dict:Dict[i]+=1
          else : Dict[i]=1

     for i in Dict:
          if i==x/2 in Dict:
               continue
          elif (x-i) in Dict:cap+=Dict[i]*Dict[x-i]

     cap/=2

     if (x/2) in Dict:cap+=Dict[x/2]*(Dict[x/2]-1)
     return int(cap)
```
