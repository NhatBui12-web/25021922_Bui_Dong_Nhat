```python
#W4A8

def deverse(a:int,so_lan=0):

    #tạo 2 list A B để so sánh
    A = [int(i) for i in str(a)]
    B = list(reversed(A))
    
 
    if A==B:
        return a,so_lan
    

    b=int("".join(str(x) for x in  B))
    a=a+b

    return deverse(a,so_lan+1 )


n=int(input())
so,so_buoc=deverse(n)
print(so_buoc,so)
```

```python
#W4A10
'''
Mô tả dãy Collatz:

Với một số nguyên dương n:

Nếu n là chẵn, thì n = n / 2

Nếu n là lẻ, thì n = 3 * n + 1

Lặp lại cho đến khi n = 1
'''
def day_collatz(n,so_lan=0):
    
    if n == 1:
        return so_lan

    if n % 2 != 0:
        n = 3 * n + 1
    else:  
        n = n // 2

    return day_collatz(n, so_lan + 1)


n=int(input())
list_so_lan=[]
n_thoa_man=[]


for i in range(1,n+1):
    so_lan=day_collatz(i)  
    list_so_lan.append(int(so_lan))

so_lan_max=max(list_so_lan)
n_thoa_man.append(list_so_lan.index(so_lan_max)+1)
print(min(n_thoa_man),so_lan_max+1)
```

```python
#W4A20
def luy_thua_2(n):
     x = 0
     while 2 ** x <= n:
          if 2 ** x == n:
            return True
          x += 1
     return False

n=int(input())
if luy_thua_2(n):
     print(True)
else:
     print(False)     
```
